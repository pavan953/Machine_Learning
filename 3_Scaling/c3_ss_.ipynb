```python
import pandas as pd
from sklearn.preprocessing import RobustScaler
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Datasets/standardscaler_ss_.csv")
df.head()
```

```python
df.info()
```

```python
sns.boxplot(df.drop(columns=["Salary"]))
```

```python
x = df.drop(columns=["Salary"])
y = df["Salary"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
scaler = RobustScaler()
```

```python
val = scaler.fit_transform(xtrain)
```

```python
sns.boxplot(val)
```

```python
val2 = scaler.transform(xtest)
```

```python
model = LinearRegression()
```

```python
model.fit(val, ytrain)
```

```python
model.score(val, ytrain)
```

```python
model.score(val2, ytest)
```
