```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = {
    "col1" : ["low", "medium", "high", "low", "high"],
    "col2" : ["red", "blue", "red", "green", "red"]
}
df = pd.DataFrame(data)
df
```

```python
encoder = OrdinalEncoder()
encoder.fit_transform(df[["col1","col2"]])
```

```python
encoder = OrdinalEncoder(categories=[["low", "medium", "high"],["red", "blue", "green"]], handle_unknown="use_encoded_value", unknown_value=-1)
encoder.fit_transform(df[["col1","col2"]])
```

```python
encoder.transform([["hii","yellow"]])
```

```python
encoder.transform([["low", "yellow"]])
```
