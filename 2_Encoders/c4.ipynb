```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = {
    "col1" : ["a", "c", "b", "c", np.nan],
    "col2" : ["x", np.nan, "z", "y", "z"]
}
df = pd.DataFrame(data)
df
```

```python
encoder = OrdinalEncoder()
encoder.fit_transform(df[["col1", "col2"]])
```
