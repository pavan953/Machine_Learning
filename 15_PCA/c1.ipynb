```python
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Datasets/creditcard_cc.csv")
df.head()
```

```python
x = df.drop(columns=["Class"])
y = df["Class"]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
pipeline = Pipeline(
    steps=[
        ("scaler", StandardScaler()),
        ("kmeans", KMeans(n_clusters=2)),
        ("model", KNeighborsClassifier())
    ]
)
```

```python
pipeline.fit(xtrain, ytrain)
```

```python
xtrain_pred = pipeline.predict(xtrain)
print(classification_report(ytrain, xtrain_pred))
```

```python
xtest_pred = pipeline.predict(xtest)
print(classification_report(ytest, xtest_pred))
```

```python
pipeline1 = Pipeline(
    steps=[
        ("scaler", StandardScaler()),
        ("kmeans", KMeans(n_clusters=2)),
        ("model", LogisticRegression())
    ]
)
```

```python
pipeline1.fit(xtrain, ytrain)
```

```python
xtrain_pred = pipeline1.predict(xtrain)
print(classification_report(ytrain, xtrain_pred))
```

```python
xtest_pred = pipeline1.predict(xtest)
print(classification_report(ytest, xtest_pred))
```
