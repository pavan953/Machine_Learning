```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Datasets/diamonds_d.csv")
df.head()
```

```python
df = df.drop(columns=["Unnamed: 0","cut","color","clarity"])
df.info()
```

```python
x = df.drop(columns=["price"])
y = df["price"]
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
gridsearchcv = GridSearchCV(
    estimator=KNeighborsRegressor(),
    param_grid={
        "n_neighbors" : [5,201,203,205,207,209,211,213],
        "weights" : ["uniform","distance"],
        "metric" : ["euclidean","manhattan"]
    },
    cv = 5,
)
```

```python
gridsearchcv.fit(xtrain,ytrain)
```

```python
gridsearchcv.best_params_
```

```python
val = gridsearchcv.best_estimator_
```

```python
val.fit(xtrain,ytrain)
```

```python
val.score(xtrain,ytrain)
```

```python
val.score(xtest,ytest)
```

```python
df1 = pd.DataFrame(gridsearchcv.cv_results_)
df1
```

```python
df1.sort_values(by="rank_test_score")
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Datasets/diamonds_d.csv")
df.head()
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.info()
```

```python
x = df.drop(columns=["price"])
y = df["price"]
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
cat_cols = xtrain.select_dtypes(include="object").columns
num_cols = xtrain.select_dtypes(include="number").columns
```

```python
for i in cat_cols:
    print(f"{i} -> {x[i].nunique()}")
```

```python
preprocessing = ColumnTransformer(
    transformers=[
        ("scaler",StandardScaler(),num_cols),
        ("encoder",OneHotEncoder(sparse_output=False, handle_unknown="ignore"),cat_cols)
    ]
)
```

```python
pipeline = Pipeline(
    steps=[
        ("preprocessing_obj", preprocessing),
        ("model", KNeighborsRegressor())
    ]
)
```

```python
gscv = GridSearchCV(
    estimator=pipeline,
    param_grid={
        "model__n_neighbors" : [5,201,203,205,207,209,211,213],
        "model__weights" : ["uniform","distance"],
        "model__metric" : ["euclidean","manhattan"]
    },
    cv = 5,
)
```

```python
gscv.fit(xtrain,ytrain)
```

```python
val2 =gscv.best_estimator_
```

```python
val2.fit(xtrain,ytrain)
```

```python
val2.score(xtrain,ytrain)
```

```python
val2.score(xtest,ytest)
```
