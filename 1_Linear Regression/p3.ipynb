```python
df['Trip_Price']=df["Trip_Price"].dropna()
df['Trip_Price'].isna().sum()
```

```python
for i in ["Time_of_Day","Day_of_Week","Weather","Traffic_Conditions"]:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
for i in ["Trip_Distance_km","Passenger_Count","Base_Fare","Per_Km_Rate","Per_Minute_Rate","Trip_Duration_Minutes"]:
    df[i] = df[i].fillna(df[i].median())
```
