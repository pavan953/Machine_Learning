```python
import pandas as pd
```

```python
data = {
    "x" : [1,2,3,4,5],
    "y" : [20,40,60,80,100]
}
df = pd.DataFrame(data)
df
```

```python
xmean = df["x"].mean()
ymean = df["y"].mean()
```

```python
df["(xi-xmean)"] = df["x"]-xmean
df["(yi-ymean)"] = df["y"]-ymean
df["(xi-xmean)(yi-ymean)"] = df["(xi-xmean)"]*df["(yi-ymean)"]
df["(xi-xmean)**2"] = df["(xi-xmean)"]**2
df
```

```python
m = df["(xi-xmean)(yi-ymean)"].sum()/df["(xi-xmean)**2"].sum()
c = ymean - m*xmean
print(f"m = {m}, c = {c}")
```
