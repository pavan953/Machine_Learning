```python
import pandas as pd
import math
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Linear Regression/Datasets/Experience-Salary.csv")
df
```

```python
df = df.rename(columns={"exp(in months)": "x","salary(in thousands)":"y"})
df["x"] = df["x"]
df["y"] = df["y"]
```

```python
xmean = df["x"].mean()
ymean = df["y"].mean()
```

```python
df["(xi-xmean)"] = df["x"] - xmean
df["(yi-ymean)"] = df["y"] - ymean
df["(xi-xmean)(yi-ymean)"] = df["(xi-xmean)"] * df["(yi-ymean)"]
df["(xi-xmean)**2"] = df["(xi-xmean)"]**2
df
```

```python
m = df["(xi-xmean)(yi-ymean)"].sum()/df["(xi-xmean)**2"].sum()
c = ymean - m * xmean
print(m,c)
```

```python
df["predicted"] = m * df["x"] + c
df["(yi-predicted)"] = df["y"] - df["predicted"]
df["(yi-predicted)**2"] = df["(yi-predicted)"]**2
df
```

```python
mse = df["(yi-predicted)**2"].sum()/df["x"].count()
mse
```

```python
rmse = math.sqrt(mse)
rmse
```

```python
mae = df["(yi-predicted)"].abs().sum()/df["x"].count()
mae
```

```python
rmae = mae / ymean
rmae
```

```python
df["(yi-ymean)**2"] = df["(yi-ymean)"]**2
```

```python
rsq_score = 1- (df["(yi-predicted)**2"].sum()/(df["(yi-ymean)**2"]).sum())
rsq_score
```
