```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Datasets/Ice_cream selling data_is.csv")
df.head()
```

```python
sns.scatterplot(df, x = "Temperature (°C)", y = "Ice Cream Sales (units)")
```

```python
df.corr()
```

__Here the relation betweem x and y is not linear so we cannot use Linear Regression__

```python
x = df[["Temperature (°C)"]]
y = df["Ice Cream Sales (units)"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtrain)
```

```python
mse = mean_squared_error(ytrain, ypred)
mse
```

```python
mae = mean_absolute_error(ytrain,ypred)
mae
```

```python
rmse = root_mean_squared_error(ytrain,ypred)
rmse
```

```python
print(f"MAE : {mae}")
print(f"RMSE : {rmse}")
```

Here the difference between mae and rmse is too low so the model prediction must be is good.

```python
r2 = r2_score(ytrain,ypred)
r2
```

But here the r2_score is soo low so the model prediction is not good.

```python
model.score(xtrain,ytrain)
```
