```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Datasets/Ice_cream selling data_is.csv")
df.head()
```

```python
x = df[["Temperature (°C)"]]
y = df["Ice Cream Sales (units)"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y , train_size=0.8, random_state=42)
```

```python
plt.scatter(xtrain,ytrain)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
pred = model.predict(xtrain)
```

```python
plt.scatter(x = xtrain, y = ytrain)
plt.scatter(x = xtrain, y = pred)
```

```python
poly = PolynomialFeatures(degree=1) # default degree = 2
xtrain_poly = poly.fit_transform(xtrain)
xtrain_poly
```

```python
model2 = LinearRegression()
model2.fit(xtrain_poly, ytrain)
```

```python
model2.score(xtrain_poly,ytrain)
```

```python
# with degree 2
poly = PolynomialFeatures(degree=2)
xtrain_poly = poly.fit_transform(xtrain)
xtrain_poly
```

```python
model3 = LinearRegression()
model3.fit(xtrain_poly, ytrain)
```

```python
model3.score(xtrain_poly,ytrain)
```

```python
ypred = model3.predict(xtrain_poly)
```

```python
xtrain_poly[:,1]
```

```python
plt.scatter(xtrain_poly[:,1],ytrain)
plt.scatter(xtrain_poly[:,1],ypred)
```

```python
val = poly.transform(xtest)
```

```python
model3.score(val,ytest)
```
