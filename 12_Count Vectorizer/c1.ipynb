```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
data = {"data" : ["machine learning","deep learning", "big data and data eng"]}
df = pd.DataFrame(data)
df
```

```python
cv = CountVectorizer(stop_words="english")
```

```python
res = cv.fit_transform(df["data"]).toarray()
```

```python
cols = cv.get_feature_names_out()
```

```python
df1 = pd.DataFrame(res, columns=cols)
df1
```

```python
cv.vocabulary_
```

```python
tfidf = TfidfVectorizer(stop_words="english")
```

```python
res = tfidf.fit_transform(df["data"]).toarray()
```

```python
cols = tfidf.get_feature_names_out()
```

```python
df2 = pd.DataFrame(res, columns=cols)
df2
```
