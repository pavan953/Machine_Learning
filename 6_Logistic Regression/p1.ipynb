```python
import numpy as np
```

```python
def precision_score(arr):
    ps_0 = arr[0,0] / (arr[0,0] + arr[1,0])
    ps_1 = arr[1,1] / (arr[1,1] + arr[0,1])
    return ps_0, ps_1

def recall_score(arr):
    rs_0 = arr[0,0] / (arr[0,0] + arr[0,1])
    rs_1 = arr[1,1] / (arr[1,1] + arr[1,0])
    return rs_0, rs_1

def accuracy_score(arr):
    acs = (arr[0,0] + arr[1,1]) / (arr[0,0] +arr[0,1] + arr[1,0] + arr[1,1])
    return acs

def f1_score(arr):
    ps_0 = arr[0,0] / (arr[0,0] + arr[1,0])
    ps_1 = arr[1,1] / (arr[1,1] + arr[0,1])
    rs_0 = arr[0,0] / (arr[0,0] + arr[0,1])
    rs_1 = arr[1,1] / (arr[1,1] + arr[1,0])
    f1s_0 = 2 * ((ps_0 * rs_0) / (ps_0 + rs_0))
    f1s_1 = 2 * ((ps_1 * rs_1) / (ps_1 + rs_1))
    return f1s_0, f1s_1
```

# Array 1

```python
arr1 = np.array([[204,193],[173,230]])
arr1
```

```python
precision_score(arr1)
```

```python
recall_score(arr1)
```

```python
accuracy_score(arr1)
```

```python
f1_score(arr1)
```
