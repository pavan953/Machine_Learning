```python
import pandas as pd
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("C:/Users/preet/Desktop/ML/Datasets/taxi_trip_pricing_ttp.csv")
df.head()
```

```python
df = df.dropna()
```

```python
x = df.drop(columns=["Trip_Price","Time_of_Day","Day_of_Week","Traffic_Conditions","Weather"])
y = df["Trip_Price"]
```

```python
xtrain, xtest, ytrain,ytest = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
model.coef_
```

```python
model_ridge = Ridge()
model_ridge.fit(xtrain, ytrain)
model_ridge.coef_
```

```python
model_lasso = Lasso()
model_lasso .fit(xtrain, ytrain)
model_lasso.coef_
```

```python
model_elasticnet = ElasticNet()
model_elasticnet.fit(xtrain, ytrain)
model_elasticnet.coef_
```
